```python
f,axes= plt.subplots(5,2,figsize=(20,30))
f.tight_layout(pad=8)  #arada 8 birimlik boşluk oluşturacak
f.suptitle("Histogram Plot")      #sub üst yazı sup alt yazı
cols = df.select_dtypes(exclude="object").columns

x_axes =0
y_axes=0


for col in cols:
  sns.histplot(data= df, x= col ,kde=True,ax=axes[x_axes,y_axes])  #kde çizgi yapacak
  if y_axes == 1:
    y_axes =0
    x_axes += 1
  else:
    y_axes +=1

plt.savefig("Plots/histogram_plots.png")
plt.show()
```

```python
f,axes= plt.subplots(5,2,figsize=(20,30))
f.tight_layout(pad=8)  #arada 8 birimlik boşluk oluşturacak
f.suptitle("Box Plot")      #sub üst yazı sup alt yazı
cols = df.select_dtypes(exclude="object").columns

x_axes =0
y_axes=0


for col in cols:
  sns.boxplot(data= df, x= col ,ax=axes[x_axes,y_axes])  #kde çizgi yapacak
  if y_axes == 1:
    y_axes =0
    x_axes += 1
  else:
    y_axes +=1

plt.savefig("Plots/box_plots.png")
plt.show()
```
